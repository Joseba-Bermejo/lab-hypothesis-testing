**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# H0 = PDragon <= HP than PGrass
# H1 = PDragon > HP than PGrass
```

```python
# filter pokemon type grass
df_grass = df[(df["Type 1"] == "Grass") | (df["Type 2"] =="Grass")]["HP"]
df_grass
```

```python
# filter pokemon type dragon
df_dragon = df[(df["Type 1"] == "Dragon") | (df["Type 2"] =="Dragon")]["HP"]
df_dragon
```

```python
# Perform t-test 
st.ttest_ind(df_dragon, df_grass, equal_var=False)
```

## Because p_value is lower than significance level (0.05%), we reject the null hypothesis, this means that dragon pokemons have higher HPs than grass pokemons.

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# H0 = LP = stas than NLP
# H1 = LP =! stats than NLP
```

```python
df.head(5)
```

```python
# filter legendary pokemon
df_legen = df[df["Legendary"] == True][["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
df_legen
```

```python
# filter non-legendary pokemon
df_no_legen = df[df["Legendary"] == False][["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
df_no_legen
```

```python
from statsmodels.multivariate.manova import MANOVA

# Combine data with a new column
df_manova = df.copy()
df_manova['Legendary'] = df_manova['Legendary'].astype(str)  # must be string/categorical

# MANOVA test
maov = MANOVA.from_formula('HP + Attack + Defense + Q("Sp. Atk") + Q("Sp. Def") + Speed ~ Legendary', data=df_manova)
print(maov.mv_test())
```

### Pr > F or p_value is extremely small, and well below 0.05. Therefore we can discard the null hypothesis and confirm that legendary pokemons have different stat values than non-legenadry ones.

```python
from scipy import stats

columns = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]

alpha = 0.05

for col in columns:
    stat_legen = df_legen[col]
    stat_nonlegen = df_no_legen[col]

    t_stat, p_val = stats.ttest_ind(stat_legen, stat_nonlegen, equal_var=False)
    
    print(f"{col}:")
    print(f"  T-statistic = {t_stat:.3f}, P-value = {p_val:.3f}")
    if p_val < alpha:
        print("  Significant difference (reject H0)")
    else:
        print("  No significant difference (fail to reject H0)")
```

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# H0 = Houses close to school or hospital (0.50) are equally or less expensive
# H1 = Houses close to school or hospital (0.50) are more expensive
```

```python
def calculate_distance(x1, x2, y1, y2):
    distance = ((x1 - x2)**2 + (y1 - y2)**2)**0.5
    return distance
```

```python
def close_or_not(df):
    distance_to_school = calculate_distance(df['longitude'], -118, df['latitude'], 34)
    distance_to_hospital = calculate_distance(df['longitude'], -122, df['latitude'], 37)

    if distance_to_school < 0.5 or distance_to_hospital < 0.5:
        return "close"
    else:
        return "far"
```

```python
df['close/far'] =  df.apply(close_or_not, axis=1)
```

```python
df['close/far'].value_counts()
```

```python
# extract median house values
close_prices = df[df['close/far'] == 'close']['median_house_value']
far_prices = df[df['close/far'] == 'far']['median_house_value']
```

```python
# perform t-test

t_stat, p_val = stats.ttest_ind(close_prices, far_prices, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

Reject H0. Therefore houses close to school or hospital are significantly more expensive.
